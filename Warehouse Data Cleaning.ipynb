```python
import seaborn as sns
import matplotlib.pyplot as plt
columns=['gender','discount_offered']
for col in columns:
    plt.figure(figsize=(10,6))
    sns.boxplot(data=df, x='gender', y='discount_offered')
    plt.xticks(rotation=45)
    plt.title("Discount Distribution by Gender")
    plt.show()
```

```python
columns=['gender','discount_offered']
for col in columns:
    plt.figure(figsize=(8,5))
    sns.countplot(data=df, x='discount_offered', order=df['gender'].value_counts().index)
    plt.xticks(rotation=45)
    plt.title("Product Category Distribution")
    plt.show()
```
